```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import math
N = 1000
d = 10
ads_selected = []
numbers_of_selections = [0] * d
sums_of_rewards = [0] * d
total_reward = 0
for n in range(0,30):
    print('start with =',n,)
    ad = 0
    max_upper_bound = 0
    for i in range(0,d):
        if (numbers_of_selections[i] > 0):
            average_reward = sums_of_rewards[i] / numbers_of_selections[i]
            delta_i = math.sqrt(3/2 * math.log(n + 1) / numbers_of_selections[i])
            upper_bound = average_reward + delta_i
            print('upper_bound  =',upper_bound)
        else:
            upper_bound = 1e400
        if upper_bound > max_upper_bound:
            max_upper_bound = upper_bound
            ad = i
            print('ad number',ad)
        print('max_upper_bound =',max_upper_bound)    
    ads_selected.append(ad)
    numbers_of_selections[ad] = numbers_of_selections[ad] + 1
    print('n & ad' ,n,ad ,sep='\t')
    reward = dataset.values[n, ad]
    sums_of_rewards[ad] = sums_of_rewards[ad] + reward
    total_reward = total_reward + reward
    
    print('reward',reward)
    print('No. of selections',numbers_of_selections)
    print('sums of rewards',sums_of_rewards)
    
    
    print(ads_selected[ad],total_reward ,sep='\t')
    
    
```

```python
import random
import pandas as pd

data = {}
data['B1'] = [random.randint(0,1) for x in range(200)]
data['B2'] = [random.randint(0,1) for x in range(200)]
data['B3'] = [random.randint(0,1) for x in range(200)]
data['B4'] = [random.randint(0,1) for x in range(200)]
data['B5'] = [random.randint(0,1) for x in range(200)]
data = pd.DataFrame(data)
```

```python
data
```

```python
observations = 200
machines = 5
```

```python
machine_selected = []
```

```python
rewards = [0] * machines
penalties = [0] * machines
total_reward = 0
```

```python
for n in range(0, observations):
    bandit = 0
    beta_max = 0
    for i in range(0, machines):
        beta_d = random.betavariate(rewards[i] + 1, penalties[i] + 1)
        print(beta_d)
        if beta_d > beta_max:
            beta_max = beta_d
            bandit = i
            print('betamax',beta_max)
    machine_selected.append(bandit)
    reward = data.values[n, bandit]
    if reward == 1:
        rewards[bandit] = rewards[bandit] + 1
    else:
        penalties[bandit] = penalties[bandit] + 1
    total_reward = total_reward + reward
```
